```python
import numpy as np
import pandas as pd
```

```python
url = r'http://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8#%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C'
```

```python
df_origin = pd.read_html(url,
                  match='Коефіцієнт народжуваності в регіонах України',
                  decimal=',',
                  thousands='.')[0]
df = df_origin.copy()
df
```

```python
# Вивести перші рядки таблиці за допомогою методу head
df.head()
```

```python
# Визначте кількість рядків та стовпців у датафреймі (атрибут shape)
df.shape
```

28 рядків, 9 стовпців

```python
# Замініть у таблиці значення "—" на значення NaN
df.replace({'—': np.nan}, inplace=True)
df
```

```python
# Визначте типи всіх стовпців за допомогою dataframe.dtypes
df.dtypes
```

```python
# Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"
df = df.astype({'2014': 'float', '2019': 'float'})
df.dtypes
```

```python
# Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)
df.isnull().sum() / df.shape[0]
```

```python
# Видаліть з таблиці дані по всій країні, останній рядок таблиці
df.drop(27, axis=0, inplace=True)
df
```

```python
# Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)
df.fillna(df.mean(numeric_only=True).round(decimals=1), inplace=True)
df
```

```python
# Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні
df[df['2019'] > df['2019'].mean()][['Регіон', '2019']]
```

```python
# У якому регіоні була найвища народжуваність у 2014 році?
df[df['2014'] == df['2014'].max()][['Регіон', '2014']]
```

```python
# Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році
df_plot = df.copy()
df_plot.index = df_plot['Регіон']
df_plot['2019'].sort_values().plot(kind='barh', figsize=(12,15), grid=True,
                     title='Коефіцієнт народжуваності в регіонах України в 2019 році')
```
