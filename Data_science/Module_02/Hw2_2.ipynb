```python
# Використовуючи метод groupby, виконайте групування за стовпчиком "Посада"
# Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", 
# виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата в місяць"
zp_df = python_data.groupby('Должность', as_index=False)
zp_df.agg({'Зарплата.в.месяц': 'max'})
```

```python
# Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. Використовуйте її для методу apply та створіть новий стовпчик "avg"
def fill_avg_salary(x):
  x = df['Зарплата.в.месяц'].mean()
  return x
```

```python
df['avg'] = df['Зарплата.в.месяц']
df['avg'] = df['avg'].apply(fill_avg_salary)
df
```
