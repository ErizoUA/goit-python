```python
import numpy as np
from scipy.interpolate import interp1d
from scipy.integrate import quad
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 1,2
speed = [25, 35, 45, 30, 60, 120, 100, 100, 70, 75, 80, 65]

time = np.linspace(0, 11, 12)
time
```

```python
# 3
plt.scatter(time, speed)
plt.xlim(0, 11)
plt.ylim(0, 130)
plt.grid(True)
```

```python
# 4

f1 = interp1d(time, speed, kind='cubic')
time_interp1 = np.linspace(0, 11, 10000)
speed_interp1 = f1(time_interp1)

plt.plot(time_interp1, speed_interp1)
plt.xlim(0, 11)
plt.ylim(0, 130)
plt.grid(True)
```

```python
# 5
integral1, _ = quad(f1, 0, 11)

integral1
```

```python
# 6
f2 = interp1d(time, speed, kind='quadratic')
time_interp2 = np.linspace(0, 11, 10000)
speed_interp2 = f2(time_interp2)

plt.plot(time_interp2, speed_interp2)
plt.xlim(0, 11)
plt.ylim(0, 130)
plt.grid(True)
```

```python
integral2, _ = quad(f2, 0, 11)

integral2
```
