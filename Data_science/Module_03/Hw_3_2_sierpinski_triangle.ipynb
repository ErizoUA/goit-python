```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Задаємо початкові координати вершин трикутника
x = np.array([0, 1, 2])
y = np.array([0, 2, 0])

# Задаємо кількість ітерацій
n = 10000

# Випадково вибираємо точку в середині трикутника
x0, y0 = np.random.uniform(0, 2), np.random.uniform(0, 2)

# Перебираємо n ітерацій
for i in range(n):
    # Випадково вибираємо вершину трикутника
    j = np.random.randint(0, 3)
    # Обчислюємо координати точки, рівновіддаленої від першої обраної точки та від обраної вершини
    x0 = (x[j] + x0) / 2
    y0 = (y[j] + y0) / 2
    # Додаємо нову точку до зображення
    plt.scatter(x0, y0, s=1)

plt.xlim(0, 2)
plt.ylim(0, 2)
plt.title("Sierpinski Triangle")
```
