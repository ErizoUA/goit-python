```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#     
x = np.array([0, 1, 2])
y = np.array([0, 2, 0])

#   
n = 10000

#      
x0, y0 = np.random.uniform(0, 2), np.random.uniform(0, 2)

#  n 
for i in range(n):
    #    
    j = np.random.randint(0, 3)
    #   ,         
    x0 = (x[j] + x0) / 2
    y0 = (y[j] + y0) / 2
    #     
    plt.scatter(x0, y0, s=1)

plt.xlim(0, 2)
plt.ylim(0, 2)
plt.title("Sierpinski Triangle")
```
